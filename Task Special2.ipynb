```python
import pandas as pd
import numpy as np
import re
```

```python
f = open('data/task/benchmark.txt')
txt = f.read()
f.close()
```

```python
myre = re.compile(r'Benchmarking (\w+) (\w+) precision type (\w+)\n(\w+)  model average (\w+) time :  (.+) ms')
myre.match(txt)
```

```python
df = pd.DataFrame(re.findall(r'Benchmarking (\w+) (\w+) precision type (\w+)\n(\w+)  model average (\w+) time :  (.+) ms', txt))
df.columns = ['type', 'precision', 'name', 'temp1', 'temp2', 'time']
df = df[['name', 'type', 'precision', 'time']]
df.head()
```

```python
df = df.pivot(index = ['name', 'type'], columns='precision', values= 'time')
df.head()
```

```python
# 平铺行
df=df.unstack()

# 对多级表头进行处理
df.columns=df.columns.map(lambda x:(x[1]+'_'+x[0]))
df = df.reset_index()
df.head()
```

```python
df.columns = ['name', 'Inference_double', 'Train_double', 'Inference_float', 'Train_float', 'Inference_half', 'Train_half']
df=df[['name','Train_half', 'Train_float', 'Train_double', 'Inference_half', 'Inference_float', 'Inference_double']]
df.head()
```

```python
df = pd.concat([df1, df2])
df['站点'] = df.MeasName.str.split('站点').apply(lambda x: x[1])
df.drop(['MeasName'],inplace=True,axis=1)
df.head()
```

```python
h = ['H0', 'H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7', 'H8', 'H9',
       'H10', 'H11', 'H12', 'H13', 'H14', 'H15', 'H16', 'H17', 'H18', 'H19',
       'H20', 'H21', 'H22', 'H23']
df=df.melt(id_vars =['Time','站点'],
             value_vars =h,var_name = 'H',
             value_name = '压力')
df.head()
```

```python
df['H']=df['H'].str.replace('H','')
df['Time'] = df['Time'] + '-' + df['H']
df['Time']=pd.to_datetime(df['Time'],format='%Y-%m-%d-%H')
df.head()
```

```python
df['H']=df['H'].str.replace('H','')
df = df[['Time', '站点', '压力']]
df.set_index('Time', inplace = True)
df.head()
```

```python
# 当天最高温、最低温和它们的温差
df3[['最高温', '最低温']] = df3['气温'].str.split('～', n=1, expand=True)

df3['最高温'] = df3['最高温'].apply(lambda x: x.strip()[:-1])
df3['最低温'] = df3['最低温'].apply(lambda x: x.strip()[:-1])

df3.head()
```

```python
df3['温差'] = pd.to_numeric(df3['最高温']) -  pd.to_numeric(df3['最低温'])
df3.drop(['最高温', '最低温'],axis = 1,  inplace = True)
df3.head()
```
