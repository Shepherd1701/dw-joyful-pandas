## 【任务一】企业收入的多样性
一个企业的产业收入多样性可以仿照信息熵的概念来定义收入熵指标（公式略），在company.csv中存有需要计算的企业和年份，在company_data.csv中存有企业、各类收入额和收入年份的信息。现请利用后一张表中的数据，在前一张表中增加一列表示该公司该年份的收入熵指标。

```python
df1 = pd.read_csv('data/task/Company.csv')
df2 = pd.read_csv('data/task/Company_data.csv')
df1.shape, df2.shape
```

```python
# 整理df2中的证券代码以及日期
df2['证券代码'] = df2['证券代码'].apply(lambda x: ('#' + ('00000' + str(x))[-6:]))
df2['日期'] = df2['日期'].apply(lambda x: x[:4])
df2.head()
```

```python
sec_nums = df1['证券代码']
sec_nums.head()
```

```python
for sec_num in set(sec_nums):
    year  = df1[df1['证券代码'] == sec_num].日期
    # 一些公司要求求多年的信息熵
    for i in year.index:
        df_temp = df2[df2['证券代码'] == sec_num]
        df_temp = df_temp[df_temp['日期'] == str(year[i])]
        total = df_temp['收入额'].sum()
        ind = df_temp['收入额'].apply(lambda x: -((x/total) * np.log(x/total))).sum()
        # 用 df1 的索引来确定位置
        df1.at[i, 'ind'] = ind
df1.head(10)
```
